```python
import numpy as np
```

```python
x=np.array([2,4,6,8,10,12])
y=np.array([3,5,10,18,48,90])
y = np.log(y)
y
```

```python
n = len(x)
sx = np.sum(x)
sy = np.sum(y)
sxy = np.sum(x*y)
sx2 = np.sum(x*x)
```

```python
A = np.array([[n, sx], [sx, sx2]])
A1 = np.array([[sy, sx], [sxy, sx2]])
A2 = np.array([[n, sy], [sx, sxy]])
```

```python
a = np.linalg.det(A1)/np.linalg.det(A)
b = np.linalg.det(A2)/np.linalg.det(A)
print(a,b)
```

```python
a= np.e ** a
a
```

```python
import matplotlib.pyplot as plt
```

```python
# Generate fitted curve
x_fit = np.linspace(min(x) - 1, max(x) + 1, 200)
y_fit = a * np.exp(b * x_fit)

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='red', label='Original Data')
plt.plot(x_fit, y_fit, color='blue', label=f'Fitted: y = {a:.2f} * e^({b:.2f}x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exponential Fit using Log Transformation and Cramer\'s Rule')
plt.legend()
plt.grid(True)
plt.show()
```
