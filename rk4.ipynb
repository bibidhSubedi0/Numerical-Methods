```python
import seaborn as sns
```

```python
def dydx(x, y):
    return x**2 - y
```

```python
def rungeKutta(x0, y0, x, h):
    plo = []
    y = y0
    for i in range(1, 10):
        k1 = h * dydx(x0, y)
        k2 = h * dydx(x0 + 0.5 * h, y + 0.5 * k1)
        k3 = h * dydx(x0 + 0.5 * h, y + 0.5 * k2)
        k4 = h * dydx(x0 + h, y + k3)
        y = y + (1.0 / 6.0)*(k1 + 2 * k2 + 2 * k3 + k4)
        plo.append(y)
        x0 = x0 + h
    return plo
```

```python
x0 = 0
y0 = 1
x1 = 2
h = 0.2
plo = rungeKutta(x0,y0,x1,h)
print(plo)
```
