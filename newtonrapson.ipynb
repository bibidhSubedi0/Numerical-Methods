```python
import math
```

```python
def func(x):
    return x**2-5*x+6
```

```python

def der(x):
    return -math.sin(x) - 1
```

```python
val= 7
error = 0.001
cntd = True
itr = 0
vals = []
while cntd and itr<25:
    val = val - (func(val)/der(val))
    cntd = abs(3-val) > error
    vals.append(val)
    itr +=1
print(abs(3-val))
print(val)
print(vals)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Plotting convergence
plt.plot(range(1, len(vals) + 1), vals, marker='o')
plt.axhline(y=3, color='r', linestyle='--', label='Root at x=3')
plt.title("Newton-Raphson Method Convergence")
plt.xlabel("Iteration")
plt.ylabel("Current Estimate of Root")
plt.legend()
plt.grid(True)
plt.show()
```
