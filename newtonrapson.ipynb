```python
def func(x):
    return x**2-5*x+6
```

```python
def der(x):
    return 2*x-5
```

```python
val= 7
error = 0.001
cntd = True
itr = 0
vals = []
while cntd and itr<10:
    val = val - (func(val)/der(val))
    cntd = abs(3-val) > error
    vals.append(val)
    itr +=1
print(abs(3-val))
print(val)
print(vals)
```
